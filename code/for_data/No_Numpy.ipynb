```python
import math
import random

def add_list(x,y):
    if (type(x) is list) & (type(y) is list):
        for i in range(len(x)):
            x[i] += y[i]
        return x

fr = open('/Users/yeseo/desktop/Selected_TaxiMach_Link_Dataset_Full_201709.txt','r')

s = fr.readline()
curr = '00'
index = '00'

sum_of_taxi_at_time = [0]*25
```
